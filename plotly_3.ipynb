```python
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmienneh "mpg"')
```

```python
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmienneh "mpg"', histnorm='probability density',
             facet_row='origin',
             color='origin')
```

```python
dff=df[(df.origin == 'usa') | (df.origin == 'europe')]
```

```python
px.histogram(dff, x='mpg', marginal='box', #marginal= 'box' dajemy wykres na marginesie
             title='Rozkład zmiennej mpg (USA vs. Europe)',
             color='origin',
             opacity=0.5)
```

```python
px.histogram(df, x='origin')
```

```python
#HISOGRAM plotly.figure_factory


import plotly.figure_factory as ff
fig = ff.create_distplot([df.mpg], group_labels=['mpg'])

fig.show()
```

```python
import numpy as np

x1= np.random.randn(5000)
x2= np.random.randn(5000) + 2

ff.create_distplot([x1, x2], ['Rozkład Normalny standardowy ~N(0,1', 'Rozkład Normalny ~N(2, 1)'],
                   curve_type='normal',
                   bin_size=0.2,
                   colors=['grey', 'magenta'])
```

```python
# Histogram plotly.graph_object histogram 2D

import plotly.graph_objects as go
fig = go.Figure(go.Histogram2d( x=df.mpg,
                               y=df.acceleration))
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(go.Histogram2d( x=df.mpg,
                               y=df.acceleration,
                               nbinsx=30,
                               nbinsy=30))

#na mapę możemy nałożyć wykres punktowy
#dodając maskę do obiektu fig

fig.add_trace(go.Scatter(
              x=df.mpg,
              y=df.acceleration,
              mode='markers',  #markers pozwala nam żeby były widoczne kropki a nie linie
              marker={        #ustawia parametry markera jeśli nie dodamy będąkropki
                  "symbol"  : 'x',     #symbol krzyżyk
                  "opacity" : 0.7,
                  "color"   : 'white',
                  "size"    :  8,
                  "line"    : {
                      'width'  :  1
                  }
              }
))

fig.show()
```

# Wykresy finansowe - OHLC
## wykresy świecowe

```python
def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df = df.reset_index()
df.head()
```

```python
df = df[df.Date > '2022-01-01']
```

```python
df.describe()
```

```python
fig = go.Figure(
    go.Ohlc(
        x=df.Date,
        open=df.Open,
        high=df.High,
        low=df.Low,
        close=df.Close
    )
)
```

```python
fig.show()
```

```python
fig = go.Figure(
    go.Candlestick(
        x=df.Date,
        open=df.Open,
        high=df.High,
        low=df.Low,
        close=df.Close
    )
)
fig.show()
```

```python
go.Waterfall()
```

```python
fig = go.Figure(go.Waterfall(
    x=['Sprzedaż stacjonarna', 'Sprzedaż Online', 'Przychody', 'Koszty', 'Podatek', 'Zysk Netto'],
    y= [60, 80, 0, -50, -30, 0],
    measure = ['relative', 'relative', 'total', 'relative', 'relative', 'total'],
    text=['+60','+80','','-50','-30'],
    textposition='outside',
    name='Rachunek zysków i strat'
))


fig.update_layout(
    title='Rachunek zysków i strat',
    showlegend=True
)
fig.show()
```

#Wykres lejkowy

Wykresy lejkowe stosujemy przede wszystkim tam gdzie mamy początkową liczbę zdarzeń a następnie zmniejsza sięna m ilość 
w tym przykładzie będzie liczba reklam liczba kliknięć i rejestracji, to może nam pomóc w ocenie np skuteczności reklamy

```python
fig = go.Figure(go.Funnel(
    x=[3500, 500, 100],
    y=['Liczba wyświetlonych reklam', 'Liczba kliknięć', 'Liczba nowych rejestracji'],
    textinfo='value+percent initial'            #ustawiamy co chcemy wyświetlić w tym przypadku wartość i procęt początkowej wartości

))

fig.show()
```

```python
fig = go.Figure()       #trochę inaczej - najpierw inicjujemy sobie obiekt pusty
fig.add_trace(go.Funnel(            #dodajemy kolejne elementy
    x=[1000, 600, 40],
    y=['Liczba wyświetlonych reklam', 'Liczba kliknięć', 'Liczba nowych rejestracji'],
    textinfo='value+percent initial',
    name='Reklama_1'
))
    
fig.add_trace(go.Funnel(
    x=[2000, 900, 160],
    y=['Liczba wyświetlonych reklam', 'Liczba kliknięć', 'Liczba nowych rejestracji'],
    textinfo='value+percent initial',
    name='Reklama_2'
))    
    
fig.add_trace(go.Funnel(
    x=[1000, 500, 50],
    y=['Liczba wyświetlonych reklam', 'Liczba kliknięć', 'Liczba nowych rejestracji'],
    textinfo='value+percent initial',
    name='Reklama_3'
))   

fig.show()
```

```python
fig = go.Figure(go.Funnelarea(
    values=[1500, 500, 100],
    text=['Liczba wyświetlonych reklam', 'Liczba kliknięć', 'Liczba nowych rejestracji']  

))

fig.show()
```
