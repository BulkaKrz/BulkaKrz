```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 20, 100)
y = np.random.rand(100, 4)
y = np.cumsum(y, axis=0)
```

```python
plt.plot(x, y)
```

```python
sns.set()
plt.plot(x, y)
```

```python
df = sns.load_dataset('tips')    # boblioteka seaborn ma kilka zbiorów danych testowych
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.describe(include=['category'])
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.set(style='darkgrid', palette='deep', font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.set(style='dark', palette='pastel', font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.set(style='whitegrid', palette='pastel', font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.set(palette='pastel', font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue= 'sex')  #parametr hue= pozwala sparametryzować 
```

```python
sns.set(palette='pastel', font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue= 'day')
```

```python
sns.set(font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', size='size')
```

```python
sns.set(font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', size='size', sizes = (10,200)) 
#parametr sizes pozwala zmienić rozmiar kropki w zależności od parametru size 
#więc 10 bęędzie odpowiadać 1 a wielkość 200 będzie odpowiadać 6
```

```python
sns.set(font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', size='size', sizes = (10,200), palette="flare", style='size') 
```

```python
sns.set(font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', palette="flare", style='sex') 
```

```python
sns.set(font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', palette="flare", style='sex', col='time') 
```

```python
sns.set(font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', palette="flare", style='sex', row='time') 
```

```python
sns.set(font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', palette="flare", style='sex', col='time', row='smoker') 
```
