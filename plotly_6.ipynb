```python
def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()

df_amzn = fetch_financial_data(company='AMZN')
df_aapl = fetch_financial_data(company='AAPL')
df_msft = fetch_financial_data(company='MSFT')
df_nflx = fetch_financial_data(company='NFLX')
```

```python
df_amzn.head(3)
```

```python
import pandas as pd

df = pd.concat([df_amzn.Close, df_aapl.Close, df_msft.Close, df_nflx.Close],
               axis=1,
               join='inner')         #nie chcemy żeby były informacje o spółkach niekompletne więc bieżemy ingormacje z dnia w którym wszystkie spółki były notowane
df.columns= ['AMAZON', 'APPLE', 'MICROSOFT', 'NETFLIX']
df= df.reset_index()
df.head()
```

```python
df.describe()
```

```python
import plotly.graph_objects as go
```

```python
trace0 = go.Scatter(x=df.Date,
                    y=df.AMAZON,
                    name='AMAZON'
                    )
trace1 = go.Scatter(x=df.Date,
                    y=df.APPLE,
                    name='APPLE'
                    )
trace2 = go.Scatter(x=df.Date,
                    y=df.MICROSOFT,
                    name='MICROSOFT'
                    )
trace3 = go.Scatter(x=df.Date,
                    y=df.NETFLIX,
                    name='NETFLIX'
                    )



data = [trace0, trace1, trace2, trace3]
layout = go.Layout(yaxis_type='log')    # ustawiamy oś y na logarytmiczną


fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace( go.Scatter(
    x= df.Date,
    y= df.AMAZON,
    name= 'AMAZON')
)
fig.add_trace( go.Scatter(
    x= df.Date,
    y= df.APPLE,
    name= 'APPLE')
)
fig.add_trace( go.Scatter(
    x= df.Date,
    y= df.MICROSOFT,
    name= 'MICROSOFT')
)
fig.add_trace( go.Scatter(
    x= df.Date,
    y= df.NETFLIX,
    name= 'NETFLIX')
)
fig.update_layout(yaxis_type='log')

fig.show()
```

# Subplots

```python
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(rows=1, cols=2, shared_yaxes=True, subplot_titles=['first', 'second'],
                    column_widths=[0.7, 0.3])
#shared_yaxes=True = ustawiamy takie same wartości na osi
fig.update_xaxes(title_text='x value', row=1, col=1)
fig.update_yaxes(title_text='y value', row=1, col=1)

fig.update_xaxes(title_text='x val', row=1, col=2)
fig.update_yaxes(title_text='y val', row=1, col=2)

fig.add_trace(
    go.Scatter(x=[1,2,3],
              y=[2,6,12]),
    row=1,
    col=1
)

fig.add_trace(
    go.Scatter(x=[1,2,3],
              y=[4,6,3]),
    row=1,
    col=2
)
fig.update_layout(title_text= ' Subplots')
fig.show()
```

```python
fig = make_subplots(rows=1, cols=2, shared_yaxes=True, subplot_titles=['first', 'second'],
                    column_widths=[0.7, 0.3], specs=[[{'type' : 'xy'}, {'type': 'domain'}]])
#shared_yaxes=True = ustawiamy takie same wartości na osi
fig.update_xaxes(title_text='x value', row=1, col=1)
fig.update_yaxes(title_text='y value', row=1, col=1)

fig.update_xaxes(title_text='x val', row=1, col=2)
fig.update_yaxes(title_text='y val', row=1, col=2)

fig.add_trace(
            go.Scatter(x=[1,2,3],
                    y=[4,6,3]),
    row=1,
    col=1
)
fig.add_trace(
    go.Pie(values=[40,30,30]),
    row=1,
    col=2
)


fig.update_layout(title_text= ' Subplots')
fig.show()
```
