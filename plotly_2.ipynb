```python
import plotly.express as px

df = px.data.gapminder()
df.head()
```

```python
px.scatter(df.query("year == 2007"), x='gdpPercap', y='lifeExp', size='pop', color='continent', log_x=True, size_max=60, height=500, title='Template {Template}')
```

```python
px.scatter(df.query("year == 2007"), x='gdpPercap', y='lifeExp', size='pop', color='continent', 
           log_x=True, size_max=60, height=500, title='Template {Template}',
           template='plotly_white')
```

```python
for template in ['plotly', 'plotly_white', 'plotly_dark', 'ggplot2', 'seaborn']:

    fig = px.scatter(df.query("year == 2007"), x='gdpPercap', 
                y='lifeExp', size='pop', color='continent', 
                log_x=True, size_max=60, height=500, title=f'Template {template}',
                template=template)

    fig.show()
```

```python

def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
```

```python
df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
df.describe()
```

```python
df = df[df['Date'] >= '2020-01-01']
df.describe()
```

```python
px.area(df, x= 'Date', y='Close')
```

```python
px.area(df, x= 'Date', y='Close', log_y=True, title='Amazon')
```

```python
pio.templates.default = 'plotly'
```

```python
from seaborn import load_dataset
import pandas as pd

df = load_dataset('flights')
df.head()
```

```python
df.describe()
```

```python
px.bar(df, x='year', y='passengers', title='Liczba pasaerw w podziale na lata',
        color='month')
```

```python
px.bar(df, x='month', y='passengers', title='Liczba pasaerw w podziale na lata', color='year')
```

```python
#Zmieniamy na wersj horyzontaln pamitaj zmie x i y !

px.bar(df, y='month', x='passengers', title='Liczba pasaerw w podziale na lata', color='year',
       orientation='h')
```

```python
px.bar(df.query('year in [1958, 1959, 1960]'), x='month', y='passengers', color='year')
```

### Wykres Ganta

```python
import plotly.figure_factory as ff
```

```python
df = [dict(Task = 'Project A', Start= '2019-01-01', Finish='2019-06-01', Complete=10),
      dict(Task = 'Project B', Start= '2019-03-01', Finish='2019-07-01', Complete=50),
      dict(Task = 'Project C', Start= '2019-02-01', Finish='2019-04-01', Complete=100),
      dict(Task = 'Project D', Start= '2019-04-01', Finish='2019-09-01', Complete=70)]
ff.create_gantt(df, colors=['#eb4034', '#34eb89'], index_col='Complete', show_colorbar=True, tasks=4)
```
